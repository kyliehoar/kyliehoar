Q9: Use the "State" column to extract all the rows that represent Massachessettes, but only keep "State," "Condition Group," "Condition," "Age Group," and "COVID-19 Deaths" columns. Create a new Pandas dataframe using these data.

```python
# answer Q9 here
# it does not matter what code you write to create the final Pandas dataframe
target_states = ['Massachusetts']
columns_to_keep = ['State', 'Condition Group', 'Condition', 'Age Group', 'COVID-19 Deaths']
filtered_df = df[df['State'].isin(target_states)]
filtered_df = filtered_df[columns_to_keep]
print(filtered_df)
```

Q10: Repeat Q9, but this time for New Hampshire, Rhode Island, Vermont, Maine and Connecticut. 

```python
# answer Q10 here
target_states = ['New Hampshire', 'Rhode Island', 'Vermont', 'Maine', 'Connecticut']
columns_to_keep = ['State', 'Condition Group', 'Condition', 'Age Group', 'COVID-19 Deaths']
filtered_df2 = df[df['State'].isin(target_states)]
filtered_df2 = filtered_df2[columns_to_keep]
print(filtered_df2)
```

Q11: Join all the dataframes from Q9 and Q10, and name it "NewEngland."

```python
# answer Q11 here
NewEngland = [filtered_df, filtered_df2]
NewEngland = pd.concat(NewEngland)
print(NewEngland)
```

Q12: Observe the specific value "All Ages" for the column "Age Group" in the NewEngland dataframe. Notice that for one state, multiple such values exist. Write code to sum the corresponding "COVID-19 Deaths" values for "All Ages" values for each state in the NewEngland dataframe. Store that information in any way you like. No specific requirements, as long as you can disply the information for each state.


```python
# Answer Q12 here
target_age = ['All Ages']
filtered_ne = NewEngland[NewEngland['Age Group'].isin(target_age)]
sum_deaths_ne = filtered_ne.groupby('State')['COVID-19 Deaths'].sum()
print(sum_deaths_ne)
```

Q13: Use Q12 to answer which of the New England states had the maximum number of COVID-19 deaths.

```python
# answer Q13 here
print('State with max COVID-19 deaths:', sum_deaths_ne.idxmax())
```
